### Extracting all boroughs that have YORK
Creating a dataframe that contains York in the text.

```python
YorkBorough_data = neighborhoods[neighborhoods.Borough.str.contains('York')]
YorkBorough_data.head()
```

### Group data

```python
#Analyze neighborhood
YorkBorough_data.groupby('Borough').count()
print('There are {} uniques categories.'.format(len(YorkBorough_data['Borough'].unique())))
```

### OneHot Encoding

```python
# ONE Hot encoding
York_data= YorkBorough_data
York_onehot = York_data[['PostCode']]
York_onehot = pd.concat([York_onehot,pd.get_dummies(York_data['Borough'])], axis=1)

York_onehot.head()
```

### Cluster Neighborhoods

```python
# set number of clusters
kclusters = 4

York_grouped_clustering = York_onehot.drop('PostCode', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(York_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```
